```python
import re

# --- Updated function to detect new WhatsApp messages ---
def is_new_message(line):
    return re.match(r"^\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{2}(?:\s| )(am|pm) - ", line)

# --- Function to extract the actual message ---
def extract_message(line):
    match = re.match(r"^\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{2}(?:\s| )(am|pm) - .*?: (.*)", line)
    if match:
        return match.group(1)
    return ""

messages = []
current_msg = ""

# --- Read the exported chat file ---
with open("clean_chat (4).txt", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if is_new_message(line):
            if current_msg:
                messages.append(current_msg.strip())
            current_msg = extract_message(line)
        else:
            current_msg += " " + line

# Add the last message
if current_msg:
    messages.append(current_msg.strip())

# --- Filter out system messages and media omitted ---
cleaned = [
    msg for msg in messages
    if msg
    and "<Media omitted>" not in msg
    and "Messages and calls are end-to-end encrypted" not in msg
    and "You created this group" not in msg
]

# --- Save to a cleaned text file ---
with open("clean_chat.txt", "w", encoding="utf-8") as f:
    for msg in cleaned:
        f.write(msg + "\n")

print("✅ Chat cleaned and saved to clean_chat.txt")
```

```python
import emoji
from collections import Counter

# Function to extract emojis from text
def extract_emojis(text):
    return [char for char in text if char in emoji.EMOJI_DATA]

# Read cleaned chat file
with open("clean_chat.txt", "r", encoding="utf-8") as f:
    all_text = f.read()

# Extract all emojis
emoji_list = extract_emojis(all_text)

# Count frequency of each emoji
emoji_counts = Counter(emoji_list)

# Get the top 5 most used emojis
top_5 = emoji_counts.most_common(5)
print("Top 5 emojis:")
for emo, count in top_5:
    print(f"{emo}: {count}")
```

```python
import emoji
from collections import Counter
import pandas as pd

# Function to extract emojis
def extract_emojis(text):
    return [char for char in text if char in emoji.EMOJI_DATA]

# Read your cleaned file
with open("clean_chat.txt", "r", encoding="utf-8") as f:
    all_text = f.read()

# Extract and count emojis
emoji_list = extract_emojis(all_text)
emoji_counter = Counter(emoji_list)

# Convert to DataFrame
df = pd.DataFrame(emoji_counter.items(), columns=["Emoji", "Count"])

# Sort by count (descending)
df = df.sort_values(by="Count", ascending=False)

# Show the result
df.head()
```

```python
df.to_csv("emoji_count.csv", index=False)

```
